## load data

```python
imgs, lbl_bbox = get_annotations('data-zaruri/_annotations.coco.json')
```

```python
img2bbox = dict(zip(imgs, lbl_bbox))
```

```python
first = {k: img2bbox[k] for k in list(img2bbox)[:1]}; first
```

```python
getters = [lambda o: Path('data-zaruri')/o, lambda o: img2bbox[o][0], lambda o: img2bbox[o][1]]
```

```python
item_tfms = [Resize(128, method='pad'),]
batch_tfms = [Rotate(), Flip(), Dihedral(), Normalize.from_stats(*imagenet_stats)]
```

```python
def get_train_imgs(noop):  return imgs
```

```python
pascal = DataBlock(blocks=(ImageBlock, BBoxBlock, BBoxLblBlock),
                 splitter=RandomSplitter(),
                 get_items=get_train_imgs,
                 getters=getters,
                 item_tfms=item_tfms,
                 batch_tfms=batch_tfms,
                 n_inp=1)
```

```python
pascal = DataBlock(blocks=(ImageBlock, BBoxBlock, BBoxLblBlock),
                 splitter=RandomSplitter(),
                 get_items=get_train_imgs,
                 get_y = [lambda o: img2bbox[o.name][0], lambda o: img2bbox[o.name][1]],
                 item_tfms=item_tfms,
                 batch_tfms=batch_tfms,
                 n_inp=1)
```

```python
dls = pascal.dataloaders('data-zaruri')
```

```python
dls.c = 20
```

```python
dls.show_batch(max_n=9)
```
